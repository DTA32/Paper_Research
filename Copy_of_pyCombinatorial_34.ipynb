```python
import pyCombinatorial
import time
```

```python
# Required Libraries
import pandas as pd

# SA
from pyCombinatorial.algorithm import simulated_annealing_tsp
from pyCombinatorial.utils import graphs, util
```

```python
# Loading Coordinates # Berlin 52 (Minimum Distance = 7544.3659)
coordinates = pd.read_csv('https://github.com/Valdecy/Datasets/raw/master/Combinatorial/TSP-02-Coordinates.txt', sep = '\t')
coordinates = coordinates.values
```

```python
# Obtaining the Distance Matrix
distance_matrix = util.build_distance_matrix(coordinates)
```

```python
# SA - Parameters
parameters = {
            'initial_temperature': 1.0,
            'temperature_iterations': 10,
            'final_temperature': 0.0001,
            'alpha': 0.9,
            'verbose': True
             }
```

```python
now = time.time()
now
```

```python
start = time.time()
print("Started at ", start)

# SA
route, distance = simulated_annealing_tsp(distance_matrix, **parameters)

end = time.time()
interval = end-start
print("Ended at ", end)
print("Program running in ", interval, "secs")
```

```python
# Plot Locations and Tour
graphs.plot_tour(coordinates, city_tour = route, view = 'notebook', size = 10)
print('Total Distance: ', round(distance, 2))
```
